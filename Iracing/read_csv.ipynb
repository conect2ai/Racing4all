```python
# Read and prepare data
# First read with low_memory=False to avoid DtypeWarning
df = pd.read_csv('data/mazda_charllote2.csv', delimiter=',', header=8, low_memory=False)

# Convert Lat and Lon to numeric, skipping the first row (units row)
df['Lat'] = pd.to_numeric(df['Lat'].iloc[1:], errors='coerce')
df['Lon'] = pd.to_numeric(df['Lon'].iloc[1:], errors='coerce')

# Drop NaN values
df_clean = df.dropna(subset=['Lat', 'Lon'])

# Create scatter plot
plt.figure(figsize=(10,8))
plt.scatter(df_clean['Lat'], df_clean['Lon'], s=10, alpha=0.5)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Track Layout')
plt.grid(True)
plt.show()
```

```python
df_clean[['Lat', 'Lon']].describe()
```

```python
# Plot distribution of Lat and Lon
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(df_clean['Lat'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Lat')
plt.ylabel('Frequency')
plt.title('Distribution of Lat')
plt.subplot(1,2,2)
plt.hist(df_clean['Lon'], bins=10, color='salmon', edgecolor='black')
plt.xlabel('Lon')
plt.ylabel('Frequency')
plt.title('Distribution of Lon')
plt.tight_layout()
plt.show()
```

```python
# Calculate min and max for better bin ranges
lat_range = df_clean['Lat'].max() - df_clean['Lat'].min()
lon_range = df_clean['Lon'].max() - df_clean['Lon'].min()

print(f'Latitude range: {lat_range:.6f} degrees')
print(f'Longitude range: {lon_range:.6f} degrees')

# Plot detailed distribution of Lat and Lon
plt.figure(figsize=(15,6))

# Latitude histogram
plt.subplot(1,2,1)
plt.hist(df_clean['Lat'], bins=50, color='skyblue', edgecolor='black')
plt.xlabel('Latitude')
plt.ylabel('Frequency')
plt.title(f'Distribution of Latitude\nRange: {df_clean["Lat"].min():.4f} to {df_clean["Lat"].max():.4f}')

# Longitude histogram
plt.subplot(1,2,2)
plt.hist(df_clean['Lon'], bins=50, color='salmon', edgecolor='black')
plt.xlabel('Longitude')
plt.ylabel('Frequency')
plt.title(f'Distribution of Longitude\nRange: {df_clean["Lon"].min():.4f} to {df_clean["Lon"].max():.4f}')

plt.tight_layout()
plt.show()
```

```python
# Create boxplots for Lat and Lon
plt.figure(figsize=(12,6))

# Create subplots
plt.subplot(1,2,1)
plt.boxplot(df_clean['Lat'], whis=1.5)
plt.title('Latitude Distribution')
plt.ylabel('Degrees')

plt.subplot(1,2,2)
plt.boxplot(df_clean['Lon'], whis=1.5)
plt.title('Longitude Distribution')
plt.ylabel('Degrees')

plt.tight_layout()
plt.show()

# Print summary statistics
print('\nLatitude Summary:')
print(df_clean['Lat'].describe())
print('\nLongitude Summary:')
print(df_clean['Lon'].describe())
```

```python
# Remove outliers using IQR method
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers from both Lat and Lon
print('Original shape:', df_clean.shape)
df_no_outliers = df_clean.copy()
df_no_outliers = remove_outliers(df_no_outliers, 'Lat')
df_no_outliers = remove_outliers(df_no_outliers, 'Lon')
print('Shape after removing outliers:', df_no_outliers.shape)

# Plot comparison of original vs cleaned data
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Original data
axes[0].scatter(df_clean['Lat'], df_clean['Lon'], s=1, alpha=0.5, color='red', label='Original')
axes[0].set_title('Original Track Layout')
axes[0].set_xlabel('Latitude')
axes[0].set_ylabel('Longitude')
axes[0].grid(True)

# Cleaned data
axes[1].scatter(df_no_outliers['Lat'], df_no_outliers['Lon'], s=1, alpha=0.5, color='blue', label='Cleaned')
axes[1].set_title('Track Layout (Outliers Removed)')
axes[1].set_xlabel('Latitude')
axes[1].set_ylabel('Longitude')
axes[1].grid(True)

plt.tight_layout()
plt.show()

# Print summary statistics of cleaned data
print('\nSummary statistics after removing outliers:')
print(df_no_outliers[['Lat', 'Lon']].describe())
```

# Analysis of Charlotte 1 Track Data
Reading and analyzing the first Charlotte track data file with the same pipeline:

```python
# Read and prepare Charlotte 1 data
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
df_ch1 = pd.read_csv('data/mazda_charllote2.csv', delimiter=',', header=8, low_memory=False)

# Convert Lat and Lon to numeric, skipping the first row (units row)
df_ch1['Lat'] = pd.to_numeric(df_ch1['Lat'].iloc[1:], errors='coerce')
df_ch1['Lon'] = pd.to_numeric(df_ch1['Lon'].iloc[1:], errors='coerce')

# Drop NaN values
df_ch1_clean = df_ch1.dropna(subset=['Lat', 'Lon'])

# Function to remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers
print('Original shape:', df_ch1_clean.shape)
df_ch1_no_outliers = df_ch1_clean.copy()
df_ch1_no_outliers = remove_outliers(df_ch1_no_outliers, 'Lat')
df_ch1_no_outliers = remove_outliers(df_ch1_no_outliers, 'Lon')
print('Shape after removing outliers:', df_ch1_no_outliers.shape)

# Create visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Scatter plot
axes[0,0].scatter(df_ch1_no_outliers['Lat'], df_ch1_no_outliers['Lon'], s=1, alpha=0.5)
axes[0,0].set_title('Track Layout (Outliers Removed)')
axes[0,0].set_xlabel('Latitude')
axes[0,0].set_ylabel('Longitude')
axes[0,0].grid(True)

# Latitude histogram
axes[0,1].hist(df_ch1_no_outliers['Lat'], bins=50, color='skyblue', edgecolor='black')
axes[0,1].set_title('Latitude Distribution')
axes[0,1].set_xlabel('Latitude')
axes[0,1].set_ylabel('Frequency')

# Longitude histogram
axes[1,0].hist(df_ch1_no_outliers['Lon'], bins=50, color='salmon', edgecolor='black')
axes[1,0].set_title('Longitude Distribution')
axes[1,0].set_xlabel('Longitude')
axes[1,0].set_ylabel('Frequency')

# Boxplots
axes[1,1].boxplot([df_ch1_no_outliers['Lat'], df_ch1_no_outliers['Lon']], labels=['Latitude', 'Longitude'])
axes[1,1].set_title('Lat/Lon Distributions')
axes[1,1].set_ylabel('Degrees')

plt.tight_layout()
plt.show()

# Print summary statistics
print('\nSummary statistics after removing outliers:')
print(df_ch1_no_outliers[['Lat', 'Lon']].describe())
```
