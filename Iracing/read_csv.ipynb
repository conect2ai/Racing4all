```python
import pandas as pd

# Replace 'iracing_vars.csv' with your desired filename if needed
df = pd.read_csv('data/mazda_charllote.csv',delimiter=',',header=8)
print(df.head())
```

```python
df_mod = df.copy()
header = df_mod.columns[:]
first_row = df_mod.iloc[0,:]
#df_mod.columns = header + ' (' + first_row + ")" #concatenate header + first row
df_mod  = df_mod.iloc[1:] #drop first row
df_mod
```

```python
# Create dictionary from column names and their units
units_dict = dict(zip(df.columns, df.iloc[0,:]))

# Print the dictionary to verify
print("Units dictionary:")
for col, unit in units_dict.items():
    print(f"{col}: {unit}")

# Save the dictionary for later use
# The dictionary can be accessed as units_dict[column_name] to get the unit
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(8,6))
plt.scatter(df_mod['Lat'], df_mod['Lon'], alpha=0.7)
plt.xlabel('Lat')
plt.ylabel('Lon')
plt.title('Scatter Plot of Lat vs Lon (Cleaned)')
plt.grid(True)
plt.show()
```

```python
# Plot distribution of Lat and Lon
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(df_mod['Lat'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Lat')
plt.ylabel('Frequency')
plt.title('Distribution of Lat')
plt.subplot(1,2,2)
plt.hist(df_mod['Lon'], bins=10, color='salmon', edgecolor='black')
plt.xlabel('Lon')
plt.ylabel('Frequency')
plt.title('Distribution of Lon')
plt.tight_layout()
plt.show()
```
