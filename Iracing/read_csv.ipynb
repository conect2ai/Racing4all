```python
# Read and prepare data
# First read with low_memory=False to avoid DtypeWarning
df = pd.read_csv('data/mazda_charllote2.csv', delimiter=',', header=8, low_memory=False)

# Convert Lat and Lon to numeric, skipping the first row (units row)
df['Lat'] = pd.to_numeric(df['Lat'].iloc[1:], errors='coerce')
df['Lon'] = pd.to_numeric(df['Lon'].iloc[1:], errors='coerce')

# Drop NaN values
df_clean = df.dropna(subset=['Lat', 'Lon'])

# Create scatter plot
plt.figure(figsize=(10,8))
plt.scatter(df_clean['Lat'], df_clean['Lon'], s=10, alpha=0.5)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Track Layout')
plt.grid(True)
plt.show()
```
